```python
import pandas as pd
import numpy as np
import random
from sklearn.preprocessing import MinMaxScaler
```

```python
file = pd.read_csv('cleaned_covid_data.csv')
date = file['date']
data_raw = file[(date == "2022-03-01")]
data_raw= data_raw.reset_index(drop = True)
data_raw = data_raw.loc[data_raw['location'] != 'United States']
data_raw = data_raw.rename(columns={"location": "country"})
```

```python
#data_raw= data_raw.drop(index=19)
data = data_raw.drop(columns=['country','date','month','year'])
scaler = MinMaxScaler()
data = scaler.fit_transform(data)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2, random_state = 0)
principalComponents = pca.fit_transform(data)
print(pca.explained_variance_ratio_)
principalDf = pd.DataFrame(data = principalComponents, columns = ['pca_dim_1', 'pca_dim_2'])
# add the result as pca_dim_1 column and pca_dim_2 column to dataset
data_raw['pca_dim_1'] = principalDf['pca_dim_1']
data_raw['pca_dim_2'] = principalDf['pca_dim_2']
```

```python
# drop US
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=265,init = 'k-means++').fit(data)
data_raw['kmeanspp_label'] = kmeans.labels_;
data_raw['kmeans_label'] = data_raw['kmeanspp_label']
```

```python
# drop US
from sklearn.cluster import MiniBatchKMeans
# fit on the whole data
kmeans = MiniBatchKMeans(n_clusters=3,random_state=0,batch_size=6,max_iter=10).fit(data)
data_raw['kmeansmini_label'] = kmeans.predict(data)
data_raw['kmeans_label'] = data_raw['kmeansmini_label']
```

```python
run visual.ipynb
```

```python
# drop US
from sklearn.cluster import DBSCAN
clustering = DBSCAN(eps=1.5, min_samples=3).fit(data)
data_raw['DBSCAN_label']  = clustering.labels_
data_raw['kmeans_label'] = data_raw['DBSCAN_label']
```

```python
run visual.ipynb
```

```python
from sklearn.mixture import GaussianMixture
gm = GaussianMixture(n_components=3, random_state=0).fit(data)
data_raw['GMM_label']  = gm.predict(data)
data_raw['kmeans_label'] = data_raw['GMM_label']
```

```python
run visual.ipynb
```

```python
from sklearn.cluster import SpectralClustering

clustering = SpectralClustering(n_clusters=3,assign_labels='discretize',random_state=0).fit(data)
data_raw['SC_label'] = clustering.labels_
data_raw['kmeans_label'] = data_raw['SC_label']
```
